```python
f = open("../data/news_apple_2022.txt", "r")
ret = f.readlines()
len(ret)
ret2 = []
date = []
for i in ret:
    if i[:5] == "2022-":
        date.append(i[:19])
        ret2.append(i[20:])
df = pd.DataFrame({'date': date,'news': ret2})
df.to_csv("../data/news_apple_2022.csv")
```

```python
f = open("../data/price_apple_2022.txt", "r")
price = f.readlines()
len(price)
price2 = []
date = []
for i in price:
    if i[:5] == "2022-":
        cur_price = i[40:]
        if cur_price[-1] == "\n":
            cur_price = cur_price[:-1]
        date.append(i[:19])
        price2.append(cur_price)
df = pd.DataFrame({'date': date,'price': price2})
df.to_csv("../data/price_apple_2022.csv")
```
